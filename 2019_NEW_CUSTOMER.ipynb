```python
import numpy as np
import pandas as pd
import os
from datetime import timedelta
from datetime import datetime
import time
pd.set_option('display.max_columns',None)
```

```python
'''默认数据路径'''
os.getcwd()
os.chdir('C:\\Users\\Administrator\\desktop\\python data  lab\\后评价数据\\2017年以来新营销客户发生不良\\')
```

```python
'''导入额度数据并进行组内排序'''
approval_data=pd.read_csv('额度数据.csv',sep='\t')
approval_data['批复日期']=pd.to_datetime(approval_data['批复日期'])
approval_data['申报机构所属一级行'].fillna('未确定申报机构所属一级行',inplace=True)
approval_data['业务性质'].fillna('未确定业务性质',inplace=True)
approval_data['sort_id']=approval_data['批复日期'].groupby(approval_data['客户编号']).rank(method='dense',ascending=True)
approval_data.sort_values(by=['客户编号','sort_id'])
```

```python
'''依据组内排序挑选新客户'''
approval_data=approval_data.loc[(approval_data['sort_id']==1)&(approval_data['批复日期']>=datetime(2017,1,1,0,0)),:]
new_cust_data=pd.DataFrame(approval_data['客户编号'].unique(),columns=['2017年以来新客户编号'])
```

```python
'''新客户对应所属一级机构'''
branch_data=approval_data.loc[:,['客户编号','申报机构所属一级行']]
branch_data=branch_data.drop_duplicates(['客户编号','申报机构所属一级行'])
branch_count=branch_data.groupby('客户编号').count()
branch_data=pd.merge(branch_data,branch_count,left_on='客户编号',right_index=True,how='left',suffixes=('','_count'))
```

```python
'''导入不良客户数据'''
cust_data=pd.read_csv('不良客户.csv',sep='\t')
cust_data['客户编号']=cust_data['客户编号'].astype('str')
```

```python
'''生成新客户发生不良数据'''
new_cust_bad_loan=pd.merge(new_cust_data,cust_data,left_on='2017年以来新客户编号',right_on='客户编号',how='inner')
new_cust_bad_loan=new_cust_bad_loan.rename(columns={'2017年以来新客户编号':'2017年以来新客户编号',
                                  '客户编号':'不良客户编号','客户名称':'不良客户名称'})
```

```python
'''生成新客户发生不良数据并匹配对应申报一级分行'''
new_cust_bad_loan=pd.merge(new_cust_bad_loan,branch_data,left_on='2017年以来新客户编号',right_on='客户编号',how='left')
new_cust_bad_loan=new_cust_bad_loan.loc[:,['2017年以来新客户编号','不良客户名称','不良客户编号','申报机构所属一级行']]
```

```python
'''生成新客户发生不良数据申报一级分行分布情况'''
bad_loan_cust_branch=new_cust_bad_loan.groupby('申报机构所属一级行').count().sort_values(by=['不良客户编号'], \
                                                                                ascending=False)
```

```python
'''结果写入对应excel'''
writer=pd.ExcelWriter( '统计结果.xlsx')
new_cust_bad_loan.to_excel(writer,'Sheet1')
bad_loan_cust_branch.to_excel(writer,'Sheet2')
writer.save()
```
